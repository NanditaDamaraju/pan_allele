```python
def get_file_from_df(df):
    new_df = pd.DataFrame(columns=('species','mhc','peptide_length','cv','sequence','inequality','meas'))
    new_df['peptide_length'] = df.Epitope.Description.str.len()
    new_df['mhc'] = df.MHC['Allele Name']
    new_df['species'] = "human"
    new_df['sequence'] = df.Epitope['Description']
    new_df['meas'] = df.Assay['Quantitative measurement']
    new_df['cv']="TDB"
    new_df['inequality']="="
    new_df = new_df.dropna()
    
    return new_df
```

```python
df_date = df[df.Reference.Date <= 2015]
new_df = get_file_from_df(df_date)
new_df.to_csv('bdata.classii.2015.csv', sep='\t', nanRep=None, index=False)
new_df
```

```python
def allelename(allele):
    bad_hla_name_elements = [
        "HLA-",
        "-",
        "*",
        ":"
    ]
    for substring in bad_hla_name_elements:
            allele = allele.replace(substring, "")
    return allele
```

```python
from sklearn.linear_model import LogisticRegression
from collections import namedtuple
```

```python
def load_binding_data(df):
    AlleleData = namedtuple("AlleleData", "Y peptides ic50")
    for allele, group in df.groupby("mhc"):
        bad_hla_name_elements = [
            "HLA-",
            "-",
            "*",
            ":"
        ]
        for substring in bad_hla_name_elements:
            allele = allele.replace(substring, "")
        ic50 = np.array(group["meas"])
        Y = np.maximum(0.0, 1.0 - np.log(ic50) / np.log(50000))
        peptides = list(group["sequence"])

        allele_groups[allele] = AlleleData(
            Y=Y,
            ic50=ic50,
            peptides=peptides)
    return allele_groups, df
```
