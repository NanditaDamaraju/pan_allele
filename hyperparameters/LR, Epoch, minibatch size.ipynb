```python
df = pd.DataFrame(columns = ('AUC', 'ACC', 'F1', 'precision', 'recall'))
with open('epoch/ffn_mult_shuffled_64_0.001') as f:
    reader = csv.reader(f,delimiter = ',')
    for rows in reader:
        metrics = ([float(x.split(')')[0]) for x in rows[1:]])
        #print pd.Series(metrics)
        df = df.append(pd.Series(metrics,index=['AUC', 'ACC', 'F1', 'precision', 'recall']),ignore_index=True )
        #print AUC, ACC, F1, precision,recall
#plt.plot(AUC, np.arange(35))
```
